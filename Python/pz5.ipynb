```python
"""
2.
Выполнить в Jupyter Lab пример сортировки выбором из
теоретической части работы (фото). Модифицировать алгоритм для сортировки
двумерного массива. Размерность массива вводиться пользователем, массив
заполняется случайными числами типа float.
"""
```

```python
import random

rows = int(input("Введите количество строк: "))
columns = int(input("Введите количество столбцов: "))

matrix = [[round(random.uniform(-100, 100), 2) for _ in range(columns)] for _ in range(rows)]

print("Исходный массив:")
for row in matrix:
    print(row)

def selection_sort_2d(matrix):
    
    for i in range(rows):
        
        for j in range(columns):
            min_i, min_j = i, j

            for x in range(i, rows):
                
                for y in range(j if x == i else 0, columns):
                    
                    if matrix[x][y] < matrix[min_i][min_j]:
                        min_i, min_j = x, y
                        
            matrix[i][j], matrix[min_i][min_j] = matrix[min_i][min_j], matrix[i][j]

selection_sort_2d(matrix)

print("Отсортированный массив:")
for row in matrix:
    print(row)
```

```python
'''
3
Выполнить в Jupyter Lab пример сортировки вставками из
теоретической части работы (§4). Модифицировать алгоритм для сортировки
двумерного массива. Размерность массива вводиться пользователем, массив
заполняется случайными числами типа float.
'''
```

```python
import random

rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

matrix = [[round(random.uniform(-100, 100), 2) for _ in range(cols)] for _ in range(rows)]

print("Исходный массив:")
for row in matrix:
    print(row)

def insertion_sort_2d(matrix):
    for i in range(rows):
        
        for j in range(cols):
            
            key = matrix[i][j]
            x, y = i, j
            
            while True:                             # двигаем элемент назад, пока он меньше предыдущих
                if y > 0:                           # есть элемент слева
                    if matrix[x][y-1] < key:
                        matrix[x][y] = matrix[x][y-1]
                        y -= 1
                    else:
                        break
                elif x > 0:                         # переходим на предыдущую строку
                    if matrix[x-1][cols-1] > key:
                        matrix[x][y] = matrix[x-1][cols-1]
                        x -= 1
                        y = cols - 1
                    else:
                        break
                else:
                    break
            matrix[x][y] = key

insertion_sort_2d(matrix)

print("Отсортированный массив:")
for row in matrix:
    print(row)
```

```python
'''
4
Выполнить в Jupyter Lab пример пирамидальной сортировки из
теоретической части работы (§5). Модифицировать алгоритм для сортировки
двумерного массива. Размерность массива вводиться пользователем, массив
заполняется случайными числами типа float.
'''
```

```python
import random

rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

matrix = [[round(random.uniform(-100, 100), 2) for _ in range(cols)] for _ in range(rows)]

print("Исходный массив:")
for row in matrix:
    print(row)

# Преобразуем двумерный массив в одномерный
nums = []
for i in range(rows):
    for j in range(cols):
        nums.append(matrix[i][j])

def heapify(nums, heap_size, root_index):
    largest = root_index
    left = 2 * root_index + 1
    right = 2 * root_index + 2

    if left < heap_size and nums[left] < nums[largest]:
        largest = left
    if right < heap_size and nums[right] > nums[largest]:
        largest = right
    if largest != root_index:
        nums[root_index], nums[largest] = nums[largest], nums[root_index]
        heapify(nums, heap_size, largest)

def heap_sort(nums):
    n = len(nums)
    for i in range(n // 2 - 1, -1, -1):
        heapify(nums, n, i)
    for i in range(n - 1, 0, -1):
        nums[i], nums[0] = nums[0], nums[i]
        heapify(nums, i, 0)

heap_sort(nums)

# Возвращаем отсортированные данные в двумерный массив
index = 0
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = nums[index]
        index += 1

print("Отсортированный массив:")
for row in matrix:
    print(row)
```

```python
'''
5.
Выполнить в Jupyter Lab пример сортировки слиянием из
теоретической части работы (§6). Модифицировать алгоритм для сортировки
двумерного массива. Размерность массива вводиться пользователем, массив
заполняется случайными числами типа float.
'''
```

```python
import random

rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

matrix = [[round(random.uniform(-100, 100), 2) for _ in range(cols)] for _ in range(rows)]

print("Исходный массив:")
for row in matrix:
    print(row)

nums = []
for i in range(rows):
    for j in range(cols):
        nums.append(matrix[i][j])

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def merge_sort(nums):
    if len(nums) <= 1:
        return nums
    mid = len(nums) // 2
    left = merge_sort(nums[:mid])
    right = merge_sort(nums[mid:])
    return merge(left, right)

nums = merge_sort(nums)

# Возвращаем отсортированные данные в двумерный массив
index = 0
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = nums[index]
        index += 1

print("Отсортированный массив:")
for row in matrix:
    print(row)
```

```python
'''
6
Выполнить в Jupyter Lab пример быстрой сортировки из
теоретической части работы (§7). Модифицировать алгоритм для сортировки
двумерного массива. Размерность массива вводиться пользователем, массив
заполняется случайными числами типа float.
'''
```

```python
import random
 
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))

matrix = [[round(random.uniform(-100, 100), 2) for _ in range(cols)] for _ in range(rows)]

print("Исходный массив:")
for row in matrix:
    print(row)

# Преобразуем двумерный массив в одномерный
nums = []
for i in range(rows):
    for j in range(cols):
        nums.append(matrix[i][j])

def partition(nums, low, high):
    pivot = nums[(low + high) // 2]
    i = low
    j = high
    while i <= j:
        while nums[i] < pivot:
            i += 1
        while nums[j] > pivot:
            j -= 1
        if i <= j:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
            j -= 1
    return i

def quick_sort(nums):
    def _quick_sort(items, low, high):
        if low < high:
            split_index = partition(items, low, high)
            _quick_sort(items, low, split_index - 1)
            _quick_sort(items, split_index, high)
    _quick_sort(nums, 0, len(nums) - 1)

quick_sort(nums)

# Возвращаем отсортированные данные в двумерный массив
index = 0
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = nums[index]
        index += 1

print("Отсортированный массив:")
for row in matrix:
    print(row)
```

```python
'''
7.
Выполнить в Jupyter Lab примеры линейного и бинарного поиска из
теоретической части работы (§8). Модифицировать алгоритмы линейного и бинарного
поиска в векторе таким образом, чтобы возвращались индексы всех вхождений
искомого элемента в вектор. Размерность вектора вводиться пользователем, вектор
заполняется случайными числами типа int
'''
```

```python
import random

n = int(input("Введите размерность вектора: "))

vector = [random.randint(0, 20) for _ in range(n)]

print("Исходный вектор:")
print(vector)

target = int(input("Введите искомый элемент: "))

def linear_search_all(lys, element):
    result = []
    for i in range(len(lys)):
        if lys[i] == element:
            result.append(i)
    return result


def binary_search_all(lys, val):
    result = []
    first = 0
    last = len(lys) - 1

    while first <= last:
        mid = (first + last) // 2
        if lys[mid] == val:
            result.append(mid)

            i = mid - 1
            while i >= 0 and lys[i] == val:
                result.append(i)
                i -= 1

            i = mid + 1
            while i < len(lys) and lys[i] == val:
                result.append(i)
                i += 1
            break
        elif val < lys[mid]:
            last = mid - 1
        else:
            first = mid + 1
    return sorted(result)


linear_result = linear_search_all(vector, target)
print(f"Линейный поиск: индексы всех вхождений элемента {target}: {linear_result}")


sorted_vector = sorted(vector)
print("Отсортированный вектор для бинарного поиска:")
print(sorted_vector)

binary_result = binary_search_all(sorted_vector, target)
print(f"Бинарный поиск: индексы всех вхождений элемента {target}: {binary_result}")
```
